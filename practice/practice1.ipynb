```python
#loading iris data
from sklearn.datasets import load_iris
data = load_iris()
```

```python
data
```

```python
data.values
```

```python
X = data.data
print(X)
```

```python
X = X[:100, [1, 3]]
print(X)
```

```python
y = data.target
```

```python
y
```

```python
y = y[:100]
```

```python
# prepare test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size = 0.2, shuffle = True, stratify = y)
```

```python
X_train
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(facecolor = 'lightgray')
plt.scatter(x = X_train[y_train == 0, 0],y = X_train[y_train == 0, 1],  marker = 'o', color = 'red')
plt.scatter(x = X_train[y_train == 1, 0],y = X_train[y_train == 1, 1],  marker = 'x', color = 'green')
```

# Logistic Regression

```python
# using Logistic Regression
from sklearn.linear_model import LogisticRegression
lr =LogisticRegression(C = 100.0, random_state = 1)
lr.fit(X_train, y_train)
```

```python
# check score of model using score function 
lr.score(X_test, y_test)
```

```python
#get coef
import numpy as np
w1, w2 = lr.coef_[0]
b = lr.intercept_[0]

x1_plot = np.linspace(X_train[:, 0].min() - 1, X_train[:, 0].max() + 1, 100)
x2_plot = -(w1/w2)*x1_plot - b/w2

#scatter data
plt.scatter(X_train[:, 0], X_train[:, 1], c = y_train, edgecolors = 'k')

#plot boundary
plt.plot(x1_plot, x2_plot)
plt.xlabel('sepal width')
plt.ylabel('petal width')
plt.show()
```

```python
import math
C_values = [0.000000000000000000000001, 0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0, 100000.0, 10000000.0]
n_cols = int(math.ceil(np.sqrt(len(C_values))))   
n_rows = int(math.ceil(len(C_values) / n_cols))
fig, axes = plt.subplots(n_rows, n_cols, figsize = (4*n_cols, 4*n_rows))
axes = axes.flatten()
for ax, C in zip(axes, C_values):
    clf = LogisticRegression(penalty = 'l2', C = C, random_state = 10, solver = 'liblinear')
    clf.fit(X_train, y_train)

    plot_decision_regions(X_train, y_train, clf = clf, ax = ax)
    ax.set_title(f'C = {C}')
plt.tight_layout()
plt.show()
```

```python
import math
C_values = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0, 100000.0, 10000000.0]
n_cols = int(math.ceil(np.sqrt(len(C_values))))   
n_rows = int(math.ceil(len(C_values) / n_cols))
fig, axes = plt.subplots(n_rows, n_cols, figsize = (4*n_cols, 4*n_rows))
axes = axes.flatten()
for ax, C in zip(axes, C_values):
    clf = SVC(C = C, random_state = 100)
    clf.fit(X_train, y_train)

    plot_decision_regions(X_train, y_train, clf = clf, ax = ax)
    ax.set_title(f'C = {C}')
plt.tight_layout()
plt.show()
```

```python
kn = ['linear', 'poly', 'rbf', 'sigmoid']
fig, axes = plt.subplots(2, 3, figsize = (8,8))
axes = axes.flatten()
for ker, ax in zip(kn, axes):
    sm = SVC(C = 100, kernel = ker, random_state = 100)
    sm.fit(X_train, y_train)
    plot_decision_regions(X_train, y_train, clf= sm, ax = ax, legend = 2)
    ax.set_title(ker)

plt.legend()
plt.show()
```

```python
#Test for gamma 
#gamma is paramter control how influence distance between points
# gamma is large then a point has small range of influence, this can lead to overfit
# gamme is small then a point has wide range of influence, this can lead to underfit
gam = [0.1, 1.0, 10.0, 100]
ncols = int(math.ceil(np.sqrt(len(gam))))
nrows = int(math.ceil(len(gam) / ncols))

fig, axes = plt.subplots(nrows, ncols, figsize = (4 * nrows, 4 * ncols))
axes = axes.flatten()
for g, ax in zip(gam, axes):
    sm = SVC(C = 100, gamma = g, random_state = 100)
    sm.fit(X_train, y_train)
    plot_decision_regions(X_train, y_train, clf= sm, ax = ax, legend = 2)
    ax.set_title(g)

plt.legend()
plt.show()
```
