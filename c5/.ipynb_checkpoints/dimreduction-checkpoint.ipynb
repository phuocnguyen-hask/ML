# Compressing Data Via Dimensionality Reduction
In this chapter, we will cover the following topics:

- **Principal Component Analysis (PCA)** for unsupervised data compression  
- **Linear Discriminant Analysis (LDA)** as a supervised dimensionality reduction technique for maximizing class separability  
- A brief overview of **nonlinear dimensionality reduction** techniques and **t-distributed stochastic neighbor embedding (t-SNE)** for data visualization


## Unsupervised dimensionality reduction via principal component analysis
**Goal:** Reduce features while retaining most information.  
**Feature Selection:** Keep original features (e.g., backward selection).  
**Feature Extraction (PCA):** Create new features (principal components).  

**Benefits:**
- Less storage & computation
- Can improve performance (reduce curse of dimensionality)

### The main steps in principal component analysis
**In a nutshell:**  
    PCA aims to find **directions of maximum variance in high-dimensional data**  
    and **projects the data into a new subspace**.


PCA for dimensionality reduction, first we construct a $ d \times k $ - dimensional matrix $ \mathbf{W}$ and the idea is using this $d \times k$ - dimensional matrix to transform a row vector in $\mathbb{R}^k$  
    to a row vector in $\mathbb{R}^d$.  
    Suppose we have a feature vector $\mathbf{x}$:

$$
\mathbf{x} = [x_1, x_2, \dots, x_d]
$$

which is then transformed by a transformation matrix $W_{d \times k}$:

$$
\mathbf{x} W = \mathbf{z}
$$

resulting in the output vector:

$$
\mathbf{z} = [z_1, z_2, \dots, z_k]
$$
- As a result of transforming original vector in $\mathbb{R}^k$ to $\mathbb{R}^d$ (typically k $\ll$  d), the first principal compenent will **have largest variance**.
- **All consequent principal component will have largest variance** given they're in constraint that they're mutually orthogonal (uncorrelated) even though input features are correlated.
- **Principal component is sensitive to data scaling**, so we should standardize the before reduce dimension of data
- **Algorithm for dimensionality reduction:**  
    1. Standardize the $d$-dimensional dataset.
    2. Construct the covariance matrix.
    3. Decompose the covariance matrix into its eigenvectors and eigenvalues.
    4. Sort the eigenvalues in decreasing order to rank the corresponding eigenvectors.
    5. Select $k$ eigenvectors corresponding to the $k$ largest eigenvalues, where $k$ is the dimensionality of the new feature subspace.
    6. Construct a projection matrix $\mathbf{W}$ from the top $k$ eigenvectors.
    7. Transform the $d$-dimensional input dataset $\mathbf{x}$ using the projection matrix $\mathbf{W}$ to obtain the new $k$-dimensional feature subspace.


### Extracting the principal component step by step
**Goal:**
- Tackling first four steps of PCA:
    1. Standardizing input
    2. Constructing covariance matrix
    3. Obtaining eigenvalues and eigenvectors of covariance matrix
    4. Sorting eigenvalues by decreasing order to rank the eigenvectors

```python
import pandas as pd

df_wine = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',
    header=None
)
```

**First step:** Standardizing data
- Just use StandardScaler from sklearn.preprocessing

```python
from sklearn.model_selection import train_test_split
X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:,0].values
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3,stratify=y, random_state=0)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.fit_transform(X_test)
```

**Second step:** Constructing covariance matrix
- Covariance matrix is a $d \times d$ matrix where $(j,k)$ position stores the covariace of feature $x_j$ and feature $x_k$ and calculated by the following fomula:
    $$
\sigma_{jk} = \frac{1}{n} \sum_{i=1}^{n} \left( x_j^{(i)} - \mu_j \right) \left( x_k^{(i)} - \mu_k \right)
$$
- **The eigenvectors of covariance matrix are principal compoments, where as eigenvalues defines their magnitude**

**Third step:** Obtaining eigenvalues and eigenvectors of covariance matrix
- Just use linear algebra knowledge, we have:
  $$ \Sigma v =  \lambda v$$
- Use linalg.eig function from NumPy to get eigenpairs

```python
import numpy as np
cov_mat = np.cov(X_train_std.T)
eigen_vals, eigen_vects = np.linalg.eig(cov_mat)
print('\nEigenvalues \n', eigen_vals)
```

### Total and explained variance
*(Also fourth step)*

The **variance explained** by *eigenvalue* is the fraction of that *eigenvalue* with sum of total *eigenvalues*. 
$$\frac{\lambda_i}{\sum_{j = 1}^n \lambda_j}$$


```python
tot = sum(eigen_vals)
var_exp = [(i / tot) for i in sorted(eigen_vals, reverse = True)]
cum_var_exp = np.cumsum(var_exp)
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2)
axes[0].plot(cum_var_exp)
axes[1].plot(var_exp)
for ax in axes:
    ax.set_xlabel('principal component index')
    ax.set_ylabel('Explained ratio')
```

### Feature transformation

```python
#Make a list of (eigenval, elgenvect) tuples
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vects[:, i]) for i in range(len(eigen_vals))]
eigen_pairs.sort(key=lambda k: k[0], reverse = True)
```

```python
eigen_vects
```

```python
w = np.hstack((eigen_pairs[0][1][:, np.newaxis],eigen_pairs[1][1][:, np.newaxis]))
print('Matrix W:\n', w)
```

**Note**: `w` is $W$ matrix, $\mathbf{x}\mathbf{W}$ is the $\mathbf{x}'$ in 2D dimension.Why $\mathbf{x}\mathbf{W}$ transform $\mathbf{x}$ into a vector in $\mathbb{R}^2$.
  - First because of shape of matrix $\mathbf{W}$
  - Second is that $\mathbf{x}\mathbf{W}=(\mathbf{W}^T\mathbf{x}^T)^T$, $\mathbf{W}^T$ is $13 \times 2$ matrix, it coverts a vector from $\mathbf{R}^{13}$ to $\mathbf{R}^2$ and most important thing is that it converts $\mathbf{e_1}$ to the *eigenvector* in $\mathbf{R}^{13}$, *etc...*

```python
X_train_std[0].dot(w)
```

```python
# transform the whole data into 2D dimension data
X_train_pca = X_train_std.dot(w)
```

```python
colors = ['r','g','b']
markers=['o','s','^']
for l,c,m in zip(np.unique(y_train),colors,markers):
    plt.scatter(X_train_pca[y_train==1,0],X_train_pca[y_train==1,1],c=c,label=f'Class{l}',marker=m)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

### Principal component analysis in scikit-learn

```python
# plot decision bounds
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    # setup marker generator and color map
    markers = ('o', 's', '^', 'v', '<')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(
        np.arange(x1_min, x1_max, resolution),
        np.arange(x2_min, x2_max, resolution)
    )
    
    lab = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    lab = lab.reshape(xx1.shape)
    plt.contourf(xx1, xx2, lab, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    # plot class examples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(
            x=X[y == cl, 0],
            y=X[y == cl, 1],
            alpha=0.8,
            c=colors[idx],
            marker=markers[idx],
            label=f'Class {cl}',
            edgecolor='black'
        )
    
    # highlight test samples
    if test_idx is not None:
        X_test, y_test = X[test_idx, :], y[test_idx]
        plt.scatter(
            X_test[:, 0],
            X_test[:, 1],
            c='',
            edgecolor='black',
            alpha=1.0,
            linewidth=1,
            marker='o',
            s=100,
            label='Test set'
        )
```

- `PCA` is another transformer class in sklearn.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
lr = LogisticRegression(random_state=1, solver='lbfgs')
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.fit_transform(X_test_std)
lr.fit(X_train_pca, y_train)
plot_decision_regions(X_train_pca, y_train, classifier = lr)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plot_decision_regions(X_test_pca,y_test,classifier=lr)
plt.xlabel('PC 1')
plt.xlabel('PC 2')
plt.legend()
plt.show()
```

```python
# access variace explained by expained_variance_ratio_ attributes
pca = PCA(n_components = None)
X_train_pca = pca.fit(X_train_std)
pca.explained_variance_ratio_
```

## Supervised data compression via linear discriminant analysis
LDA can be used to extract feature, increse computation, reduce degree of overfitting due to the curse of dimensionality, unlike PCA, LDA don't find eigenvectors but find feature subspace that optimizes outcome

### PCA versus LDA

  
    A linear discriminant in x-axis (LD1), would separate the two normal distributions classes as well, although LD2 captures almost variability of sample, it doesn't serve as a good indicator because **it doesn't capture information about class-discriminatory**  
    One assumption in LDA is that data is normally distributed, and we aksi assume that the classes have  
    identical covariance matrices and that the training examples are statistically independent of each other.  
    If one of these assumptions are violated, performance is affected, but in some cases it's still good classifier. 

### The inner workings of LDA

**Steps to Perform Linear Discriminant Analysis (LDA)**

Linear Discriminant Analysis (LDA) is similar to PCA, but it uses class labels. The steps are:

1. **Standardize the dataset**  
   Standardize the $d$-dimensional dataset ($d$ = number of features).

2. **Compute class mean vectors**  
   For each class, compute the $d$-dimensional mean vector $\mathbf{m}_i$.

3. **Construct scatter matrices**  
   - **Between-class scatter matrix** $S_B$  
   - **Within-class scatter matrix** $S_W$

4. **Compute eigenvectors and eigenvalues**  
   Solve the eigenproblem for $S_W^{-1} S_B$.

5. **Sort eigenvectors**  
   Sort the eigenvalues in decreasing order to rank the eigenvectors.

6. **Select top $k$ eigenvectors**  
   Choose the $k$ eigenvectors corresponding to the $k$ largest eigenvalues to form a $d \times k$ transformation matrix $W$. The eigenvectors are the columns of $W$.

7. **Project data onto the new subspace**  
   Project the examples using the transformation matrix $W$:
   $$
   \mathbf{x}' = \mathbf{x} W
   $$

**Note:** LDA is supervised (uses class labels), unlike PCA, but both rely on eigen decomposition to create a lower-dimensional feature space.


### Computing the scatter matrix

```python
#print mean of features of classes
np.set_printoptions(precision=4)
mean_vecs=[]
for label in range(1,4):
    mean_vecs.append(np.mean(X_train_std[y_train==label],axis=0))
    print(f'MV {label}: {mean_vecs[label-1]}\n')
```

Using the mean vectors, we can compute with-in class scatter matrix, $\mathbf{S_w}$:
$$\mathbb{S_W} = \sum_{i=1}^{c}S_i$$
$\mathbb{S_i}$ is calculated by formula:
$$\mathbb{S_i} = \sum_{x \in D_i}(x - m_i)(x - m_i)^T$$

```python
d = 13 # #features
S_W = np.zeros((d,d))
for label, mv in zip(range(1,4), mean_vecs):
    class_scatter = np.zeros((d,d))
    for row in X_train_std[y_train==label]:
        row, mv = row.reshape(d, 1), mv.reshape(d, 1)
        class_scatter += (row - mv).dot((row-mv).T)
    S_W += class_scatter
```

$\mathbb{S_i}$ formula is similar to the formula for calculating the covariance matrix in the [Extracting the principal component](#extracting-the-principal-component-step-by-step) section.
Indeed $$\Sigma_i = \frac{1}{n_i}S_i = \frac{1}{n_i} \sum_{x \in D_i} (x - m_i)(x - m_i)^T$$
Then $$\mathbb{S_B} = \sum_{i=1}^c n_i(m_i - m)(m_i - m)^T$$


```python
#Calculate scaled version using formulat above
S_W = np.zeros((d,d))
for label, mv in zip(range(1,4), mean_vecs):
    class_scatter = np.cov(X_train_std[y_train==label].T)
    S_W += class_scatter
```

```python
import numpy as np

# Number of features
d = 13

# Compute overall mean
mean_overall = np.mean(X_train_std, axis=0).reshape(d, 1)

# Initialize between-class scatter matrix
S_B = np.zeros((d, d))

# Compute between-class scatter matrix
for i, mean_vec in enumerate(mean_vecs):
    n = X_train_std[y_train == i + 1, :].shape[0]
    mean_vec = mean_vec.reshape(d, 1)  # make column vector
    S_B += n * (mean_vec - mean_overall).dot((mean_vec - mean_overall).T)
```

### Selecting linear discriminants for the new  feature  subspace

The remaining steps of LDA is similar to CPA, but instead of find eigenvalues of covariance matrix, we finding eigenvalues of $\mathbb{S_W}^{-1}\mathbb{S_b}$

```python
eigen_vals, eigen_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))
```

```python
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:,i]) for i in range(len(eigen_vals))]
eigen_pairs = sorted(eigen_pairs, key = lambda k: k[0], reverse = True)
```

```python
for eigen_val in eigen_pairs:
    print(eigen_val[0])
    print('\n')
```

```python
#how much of the class-disciminatory is allocated via eigenvecs
tot = sum(eigen_vals.real)

# Individual discriminability (sorted by descending eigenvalues)
discr = [i / tot for i in sorted(eigen_vals.real, reverse=True)]

# Cumulative discriminability
cum_discr = np.cumsum(discr)

# Plot
plt.bar(range(1, 14), discr, align='center', label='Individual discriminability')
plt.step(range(1, 14), cum_discr, where='mid', label='Cumulative discriminability')
plt.ylabel('"Discriminability" ratio')
plt.xlabel('Linear Discriminants')
plt.ylim([-0.1, 1.1])
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# do the as i did in PCA
w = np.hstack((eigen_pairs[0][1][:,np.newaxis].real, eigen_pairs[1][1][:,np.newaxis].real))
print(w)
```

### Project examples onto new space

It's just the same as PCA

```python
X_train_lda = X_train_std.dot(w)

# Colors and markers for classes
colors = ['r', 'b', 'g']
markers = ['o', 's', '^']

# Scatter plot of LDA-transformed data
for l, c, m in zip(np.unique(y_train), colors, markers):
    plt.scatter(
        X_train_lda[y_train == l, 0],
        X_train_lda[y_train == l, 1] * (-1),  # Flip second component if desired
        c=c,
        label=f'Class {l}',
        marker=m
    )

plt.xlabel('LD 1')
plt.ylabel('LD 2')
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
